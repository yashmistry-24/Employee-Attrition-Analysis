```python
data = data.drop(['EmployeeID','Age','BusinessTravel','DistanceFromHome','Education','EducationField','EmployeeCount','Gender','MaritalStatus','NumCompaniesWorked','Over18','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsWithCurrManager'], axis='columns')
data.head()
```

```python
# Handling missing values
# ['EnvirommentSatisfaction','JobSatisfaction','WorkLifeBalance']

data['EnvironmentSatisfaction'] = data['EnvironmentSatisfaction'].fillna(data['EnvironmentSatisfaction'].mean())
data['JobSatisfaction'] = data['JobSatisfaction'].fillna(data['JobSatisfaction'].mean())
data['WorkLifeBalance'] = data['WorkLifeBalance'].fillna(data['WorkLifeBalance'].mean())
data.head()
```

```python
# Creating a new feature named 'Salary' ('high' if MonthlyIncome>100000 'medium' if MonthlyIncome>50000<100000 else 'low')
def categorize_salary(income):
    if income > 100000:
        return 'high'
    elif 50000 < income <= 100000:
        return 'medium'
    else:
        return 'low'
    
data['Salary'] = data['MonthlyIncome'].apply(categorize_salary)
data
```

```python
pd.crosstab(data.Salary,data.Attrition).plot(kind='bar')
```

```python
# Above bar chart shows employees with high salaries are likely to not leave the company
```

```python
# Visualizing department wise attrition
pd.crosstab(data.Department,data.Attrition).plot(kind='bar')
```

```python
data.to_csv('Cleaned data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
attrition_encoder = LabelEncoder()
department_encoder = LabelEncoder()
jobrole_encoder = LabelEncoder()
data['Attrition'] = attrition_encoder.fit_transform(data['Attrition'])
data['Department'] = department_encoder.fit_transform(data['Department'])
data['JobRole'] = jobrole_encoder.fit_transform(data['Attrition'])
```

```python
data.head()
```

```python
salary_dummies = pd.get_dummies(data.Salary, prefix='salary')
salary_dummies.head()
```

```python
newdata_with_dummies = pd.concat([data,salary_dummies],axis='columns')
newdata_with_dummies.head()
```

```python
newdata_with_dummies.drop(['Salary','Attrition'],axis='columns',inplace=True)
newdata_with_dummies.head()
```

```python
X = newdata_with_dummies
y = data.Attrition
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
